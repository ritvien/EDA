# Cereal products

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

```python
c_df =  pd.read_csv('data_cereal.csv')
c_df
```

```python
#irrelevant fields
fields=['shelf','rating','weight','cups']
c_df=c_df.drop(fields,axis = 1)
```

```python
#corr() : mối tương quan giữa các dữ liệu
c_corr = c_df.corr()
c_corr
```

```python
sns.heatmap(data=c_corr,cmap='Blues',linecolor='white',linewidths=0.5)
```

```python
# ones_like : build a matrix of booleans(True,False) with the same shape as our dât
ones_corr = np.ones_like(c_corr,dtype=bool)
ones_corr.shape
```

```python
# np's triu: return only upper triangle matrix
mask = np.triu(ones_corr)
mask
```

```python
sns.heatmap(data=c_corr,cmap='Blues',linecolor='white',linewidths=0.5,mask=mask)
```

```python
# xóa hàng 1 cột 1
adjusted_mask=mask[1:, :-1]
adjusted_c_corr = c_corr.iloc[1: , :-1]
adjusted_c_corr
```

```python
fig,ax = plt.subplots(figsize = (10,8))
sns.heatmap(data=adjusted_c_corr,cmap='Blues',linecolor='white',linewidths=0.5,mask=adjusted_mask,annot=True,fmt=".2f")

#VIẾT HOA
yticks = [i.upper() for i in adjusted_c_corr.index]
xticks = [i.upper() for i in adjusted_c_corr.columns]
ax.set_yticklabels(yticks);
ax.set_xticklabels(xticks,rotation=90);
```

# BOX PLOT  (Distribution): 
## Sử dụng để hiển thị PHÂN PHỐI và TƯƠNG QUAN của dữ liệu số trong một tập dữ liệu.


## Chú thích:
#### TRONG HỘP: 
* Đáy hộp(Q1) là phân vị thứ 25 : là giá trị ở vị trí (n+1)/4 
* Gạch ở giữa (Q2)là trung vị: giá trị ở giữa tập dữ liệu
* Đỉnh hộp (Q3) là phân vị thứ 75: là giá trị ở vị trí 3(n+1)/4
#### 2 SỪNG:
* IQR = Q3-Q1
* Sừng ở dưới > Q1-1.5xIQR
* Sừng ở trên < Q3+1.5xIQR
#### CÁC GIÁ TRỊ NGOÀI KHOẢNG 2 SỪNG ĐC COI LÀ GIÁ TRỊ NGOẠI LỆ (OUTLIERS)

```python
sns.boxplot(data=iris_df,x='species',y='sepal_length');
```

#### Setosa
* q1 = 4.8 , q2 = 5 , q3 = 5.2 (ở trên)
* IQR = 0.4
* gạch dưới > 4.8 - 1.5x0.4 = 4.2 => 4.3
* gạch trên < 5.2 + 1.5x0.4 = 5.8 => 5.8
#### versicolor
* q1 = 5.6, q2 = 5.9, q3 = 6.3
* IQR = 0.7
* gạch dưới > 4.55 = 4.9
* gạch trên < 7.35 = 7
#### virginica
* q1 = 6.225, q2 = 6.5, q3 = 6.9
* IQR = 0.675
* gạch dưới > 5.2125 = 5.6
* gạch trên < 7.9125 = 7.9

# HISTOGRAM (distribution): Biểu đồ thổng kê PHÂN PHỐI tần suất của dữ liệu
## biểu đồ này chia thành các khoảng(bins) và đếm số lượng quan sát nằm trong mỗi khoảng
### Công dụng: 
* Cung cấp hình ảnh tổng quan về phân phối dữ liệu một cách dễ hiểu và nhanh chóng
* Cho thấy các sự biến đổi trong dữ liệu như: tập trung, xu hướng, đỉnh, độ lệch chuẩn
* So sánh phân phối của các dữ liệu khác nhau trên cùng 1 bảng

## Độ lệch chuẩn:
* Tính gt trung bình: numpy.mean()
* Tính độ sai lệch(Deviation): Khoảng các giữa mỗi điểm dữ liệu và gt trung bình.
* Tính Bình Phương của độ sai lệch: numpy.square() hoặc **
* Tính trung bình của bp độ sai lệch: Phương sai
* Tính căn bậc 2 của phương sai: numpy.sqrt()
#### hàm tính độ lệch chuẩn: numpy.std()

* Phát hiện sự biến thiên trong dữ liệu, bao gồm đỉnh và khuynh hướng của mật độ
### => Hiểu sâu hơn về phân phối dữ liệu và quan sát sự biến thiên tại mức độ khác nhau so với boxplot

## KHẮC PHỤC:

```python
sns.histplot(data = df,x='x',y='y');
```

* Marginal Distribution (Phân phối biên)
